## Handling missing value

```python
df.Dependents.value_counts()
```

```python
df.Self_Employed.value_counts()
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna("No")
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df.LoanAmount.mean())
```

```python
df.Loan_Amount_Term = df.Loan_Amount_Term.fillna(df.Loan_Amount_Term.mean())
```

```python
df.Credit_History.value_counts()
```

```python
df.Credit_History = df.Credit_History.fillna(1.0)
```

# One hand encoding
### make our catagorical data computer understandable

```python
df.dtypes
```

```python
for i ,c in enumerate(df.dtypes):
    print(i)
    print(c)
    print()
```

```python
# Lets separate the categorical and numerical columns 

cat_data = []
num_data = []

for i,c in enumerate(df.dtypes):
    if c == object:
        cat_data.append(df.iloc[:, i])
    else:
        num_data.append(df.iloc[:, i])
```

```python
#our seperated datasets are in the form of list 
# cat_data.head()
# to check the errors you can run above code
```

```python
# now lets convert them into dataframe using pandas for fourther process
cat_data = pd.DataFrame(cat_data)
num_data = pd.DataFrame(num_data)
```

```python
cat_data.head()
```

```python
num_data.head()
```

### it's not in the form we want 
##### so lets transpose rows into columns and vise varsa 

```python
cat_data = pd.DataFrame(cat_data).transpose()
num_data = pd.DataFrame(num_data).transpose()
```

```python
cat_data.head()
```

```python
num_data.head()
```

```python
cat_data.drop('Loan_ID', axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# now we can transform the catagorical datasets into numerical datasets 
# we can also use get_dummies function to apply same feature 
for i in cat_data:
    cat_data[i] = le.fit_transform(cat_data[i])
```

```python
cat_data.head()
```

```python
## mappint 1.0 to 0 and 0.0 to 0 
mapping = {
    1.0 : 1,
    0.0 : 0
}
```

```python
num_data.loc[ : , 'Credit_History'] = num_data['Credit_History'].map(mapping)
```

```python
num_data.head()
```

```python
## now lest concat both data into a single dataframe

df = pd.concat([cat_data , num_data] , axis = 1)
```

```python
df.head()
```

```python
df.shape
```

## Training the dataframe

```python
X = df.drop('Loan_Status', axis = 1)
y = df["Loan_Status"]
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.3 , random_state = 42) 
```

```python
print('X_train shape', X_train.shape)
print('y_train shape', y_train.shape)
print('X_test shape', X_test.shape)
print('y_test shape', y_test.shape)
```

# model part

```python

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier


models = {
    'LogisticRegression' : LogisticRegression(random_state = 42),
    'KNeighborsClassifier' : KNeighborsClassifier(),
    'SVC' : SVC(random_state = 42),
    'DecisionTreeClassifier' : DecisionTreeClassifier(max_depth = 5, random_state = 42)
}
# Building the Functions

from sklearn.metrics import precision_score, recall_score, f1_score, log_loss, accuracy_score

def loss(y_true, y_pred, retu=False):
    pre = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    loss = log_loss(y_true, y_pred)
    acc = accuracy_score(y_true, y_pred)
    
    if retu:
        return pre, rec, f1, loss, acc
    else:
        print('  pre: %.4f\n  rec: %.4f\n  f1: %.4f\n  loss: %.4f\n  acc: %.4f' % (pre, rec, f1, loss, acc))
# Evaluating the model

def train_eval(models, X, y):
    for name, model in models.items():
        print(name, ':')
        model.fit(X, y)
        loss(y, model.predict(X))
        print('#'*40)
train_eval(models, X_train, y_train)
```
